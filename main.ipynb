### Feature Selection Part

```python
data.columns
```

```python
data = data[['id','title','genre','overview']]
data
```

```python
data['tag'] = data['overview'] + data['genre']
```

```python
data.head(2)
```

```python
movie = data.drop(columns=['overview','genre'])
movie.head(5)
```

# 1.Bag of word
# 2.tfidf

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=10000,stop_words='english')
cv
```

```python
vector = cv.fit_transform(movie['tag'].values.astype('U')).toarray()
```

```python
vector.shape
```

# Cosine similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
# from scipy.sparse import csr_matrix
```

```python
# vector = csr_matrix(vector.shape, dtype=np.float32)

similarity = cosine_similarity(vector)
# similarity
```

```python
def recommand(movies):
    index = movie[movie['title']==movies].index[0]
    distance = sorted(list(enumerate(similarity[index])),reverse=True,key=lambda vector:vector[1])
    for i in distance[0:5]:
        print(movie.iloc[i[0]].title)
```

```python
recommand('Iron Man')
```
